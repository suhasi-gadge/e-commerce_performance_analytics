---## 2. Data Cleaning Pipeline <a id="2-cleaning"></a>Replicating the SQL cleaning logic in Python for reproducibility. Each step is documented with the issue found and decision made.

### 2.1 Quick Data Quality Scan

```python
# Null count summary across all tablesfor name, df in [('orders', orders), ('items', items), ('customers', customers),                  ('payments', payments), ('reviews', reviews), ('products', products)]:    nulls = df.isnull().sum()    nulls = nulls[nulls > 0]    if len(nulls) > 0:        print(f"\n{name.upper()} — null columns:")        for col, count in nulls.items():            print(f"  {col:40s} → {count:,} nulls ({100*count/len(df):.1f}%)")    else:        print(f"\n{name.upper()} — no nulls ✓")
```

### 2.2 Fix Category Translations

```python
# Add the 2 missing category translationsextra_translations = pd.DataFrame({    'product_category_name': ['pc_gamer', 'portateis_cozinha_e_preparadores_de_alimentos'],    'product_category_name_english': ['pc_gaming', 'portable_kitchen_food_processors']})translation = pd.concat([translation, extra_translations], ignore_index=True)print(f"Translation table: {len(translation)} categories (was 71, added 2)")
```

### 2.3 Deduplicate Reviews**Issue:** 814 duplicate review_ids, 547 orders with multiple reviews.  **Decision:** Keep the most recent review per order.

```python
# Beforeprint(f"Before dedup: {len(reviews):,} reviews, {reviews['order_id'].nunique():,} unique orders")print(f"Duplicate review_ids: {reviews['review_id'].duplicated().sum()}")print(f"Orders with multiple reviews: {reviews.groupby('order_id').size().gt(1).sum()}")# Deduplicatereviews['review_answer_timestamp'] = pd.to_datetime(reviews['review_answer_timestamp'])reviews_clean = (reviews    .sort_values(['order_id', 'review_answer_timestamp', 'review_score'], ascending=[True, False, False])    .drop_duplicates(subset='order_id', keep='first'))reviews_clean['has_comment'] = reviews_clean['review_comment_message'].notna()print(f"\nAfter dedup: {len(reviews_clean):,} reviews (1:1 with orders) ✓")
```

### 2.4 Clean Products

```python
# Join translations and handle nullsproducts_clean = products.merge(translation, on='product_category_name', how='left')products_clean['product_category_name'] = products_clean['product_category_name'].fillna('unknown')products_clean['product_category_name_english'] = products_clean['product_category_name_english'].fillna('unknown')# Rename typo columnsproducts_clean = products_clean.rename(columns={    'product_name_lenght': 'product_name_length',    'product_description_lenght': 'product_description_length'})# Add derived columnsproducts_clean['product_volume_cm3'] = (    products_clean['product_length_cm'] * products_clean['product_height_cm'] * products_clean['product_width_cm'])products_clean['product_size_tier'] = pd.cut(    products_clean['product_weight_g'],    bins=[0, 500, 2000, 10000, float('inf')],    labels=['small', 'medium', 'large', 'extra_large'])print(f"Products cleaned: {len(products_clean):,}")print(f"Categories with English names: {(products_clean['product_category_name_english'] != 'unknown').sum():,} / {len(products_clean):,}")
```

### 2.5 Clean Orders — Cast Dates & Derive Metrics

```python
# Cast date columnsdate_cols = ['order_purchase_timestamp', 'order_approved_at', 'order_delivered_carrier_date',             'order_delivered_customer_date', 'order_estimated_delivery_date']for col in date_cols:    orders[col] = pd.to_datetime(orders[col])# Derive calendar dimensionsorders['order_month'] = orders['order_purchase_timestamp'].dt.to_period('M').dt.to_timestamp()orders['order_year']  = orders['order_purchase_timestamp'].dt.yearorders['order_quarter'] = orders['order_purchase_timestamp'].dt.quarterorders['order_dow']   = orders['order_purchase_timestamp'].dt.dayofweekorders['order_hour']  = orders['order_purchase_timestamp'].dt.hour# Derive delivery metrics (delivered orders only)mask = (orders['order_status'] == 'delivered') & orders['order_delivered_customer_date'].notna()orders['delivery_days'] = np.where(mask,    (orders['order_delivered_customer_date'] - orders['order_purchase_timestamp']).dt.total_seconds() / 86400, np.nan)orders['handling_days'] = np.where(orders['order_delivered_carrier_date'].notna(),    (orders['order_delivered_carrier_date'] - orders['order_purchase_timestamp']).dt.total_seconds() / 86400, np.nan)orders['transit_days'] = np.where(mask & orders['order_delivered_carrier_date'].notna(),    (orders['order_delivered_customer_date'] - orders['order_delivered_carrier_date']).dt.total_seconds() / 86400, np.nan)orders['delay_days'] = np.where(mask,    (orders['order_delivered_customer_date'] - orders['order_estimated_delivery_date']).dt.total_seconds() / 86400, np.nan)orders['is_late'] = mask & (orders['order_delivered_customer_date'] > orders['order_estimated_delivery_date'])orders['estimated_delivery_days'] = np.where(mask,    (orders['order_estimated_delivery_date'] - orders['order_purchase_timestamp']).dt.total_seconds() / 86400, np.nan)print(f"Orders with delivery metrics: {mask.sum():,} / {len(orders):,}")print(f"Late deliveries: {orders['is_late'].sum():,} ({100*orders['is_late'].mean():.1f}%)")
```

### 2.6 Filter to Analysis Window & Build Master DataFrames

```python
# Filter to complete months: Jan 2017 – Aug 2018orders_f = orders[    (orders['order_purchase_timestamp'] >= '2017-01-01') &    (orders['order_purchase_timestamp'] < '2018-09-01')].copy()print(f"Orders in analysis window: {len(orders_f):,} (of {len(orders):,} total)")# Clean paymentspayments_clean = payments[    (payments['payment_value'] > 0) & (payments['payment_type'] != 'not_defined')].copy()# Aggregate payments to order levelpay_agg = (payments_clean    .sort_values('payment_value', ascending=False)    .drop_duplicates('order_id', keep='first')    [['order_id', 'payment_type', 'payment_installments']]    .rename(columns={'payment_type': 'primary_payment_type', 'payment_installments': 'installments'}))# Build item-level master dataframedf = (orders_f    .merge(items, on='order_id', how='inner')    .merge(customers, on='customer_id', how='inner')    .merge(products_clean[['product_id','product_category_name_english','product_weight_g',                           'product_volume_cm3','product_size_tier','product_photos_qty',                           'product_description_length']], on='product_id', how='left')    .merge(sellers[['seller_id','seller_state','seller_city']], on='seller_id', how='left')    .merge(reviews_clean[['order_id','review_score','has_comment','review_comment_message']], on='order_id', how='left'))df['total_item_value'] = df['price'] + df['freight_value']# Build order-level summaryorder_agg = df.groupby('order_id').agg(    order_status=('order_status','first'),    order_purchase_timestamp=('order_purchase_timestamp','first'),    order_month=('order_month','first'),    order_year=('order_year','first'),    order_quarter=('order_quarter','first'),    order_dow=('order_dow','first'),    order_hour=('order_hour','first'),    delivery_days=('delivery_days','first'),    handling_days=('handling_days','first'),    transit_days=('transit_days','first'),    delay_days=('delay_days','first'),    is_late=('is_late','first'),    estimated_delivery_days=('estimated_delivery_days','first'),    customer_state=('customer_state','first'),    customer_unique_id=('customer_unique_id','first'),    seller_state=('seller_state','first'),    item_count=('order_item_id','count'),    order_revenue=('price','sum'),    order_freight=('freight_value','sum'),    review_score=('review_score','first'),    has_comment=('has_comment','first'),).reset_index()order_agg['order_total'] = order_agg['order_revenue'] + order_agg['order_freight']order_agg['freight_pct'] = 100 * order_agg['order_freight'] / order_agg['order_revenue'].replace(0, np.nan)# Merge payment infoorder_agg = order_agg.merge(pay_agg, on='order_id', how='left')print(f"\nMaster item-level  : {len(df):,} rows")print(f"Master order-level : {len(order_agg):,} rows")print(f"Date range         : {order_agg['order_purchase_timestamp'].min().date()} to {order_agg['order_purchase_timestamp'].max().date()}")
```

### 5.2 Delivery Decomposition: Handling vs TransitThis is a key diagnostic — it separates **seller responsibility** (handling: order to carrier) from **carrier responsibility** (transit: carrier to customer).

---## 7. Cross-Category: Statistical Deep Dive <a id="7-crosscat"></a>This section connects findings across all four categories with statistical tests.

### 7.1 Correlation Matrix: What Drives Review Scores?

```python
# Build correlation datasetcorr_df = rev_del[['review_score','delivery_days','handling_days','transit_days',                    'delay_days','order_revenue','order_freight','freight_pct','item_count']].dropna()# Spearman correlations with review scoreprint("Spearman Correlations with Review Score:")print("-" * 50)for col in ['delivery_days','handling_days','transit_days','delay_days',            'order_revenue','order_freight','freight_pct','item_count']:    r, p = stats.spearmanr(corr_df[col], corr_df['review_score'])    sig = '***' if p < 0.001 else '**' if p < 0.01 else '*' if p < 0.05 else ''    print(f"  {col:20s}  ρ = {r:+.3f}  p = {p:.2e}  {sig}")# Heatmapfig, ax = plt.subplots(figsize=(10, 8))corr_matrix = corr_df.corr(method='spearman')mask = np.triu(np.ones_like(corr_matrix, dtype=bool))sns.heatmap(corr_matrix, mask=mask, annot=True, fmt='.2f', cmap='RdBu_r', center=0,            vmin=-0.5, vmax=0.5, ax=ax, square=True, linewidths=0.5)ax.set_title('Spearman Correlation Matrix')plt.tight_layout()plt.savefig('../images/14_correlation_matrix.png', dpi=150, bbox_inches='tight')plt.show()
```

### 7.2 Multi-Factor Analysis: What Causes 1-Star Reviews?

```python
# Multi-factor crosstabrev_del['late_label'] = np.where(rev_del['is_late'], 'Late', 'On Time')rev_del['freight_label'] = np.where(rev_del['freight_pct']>30, 'High Freight', 'Normal Freight')multi = rev_del.groupby(['late_label','freight_label']).agg(    orders=('order_id','count'),    avg_score=('review_score','mean'),    pct_1star=('review_score', lambda x: 100*(x==1).mean())).reset_index()print("Multi-Factor Satisfaction Analysis:")print(multi.to_string(index=False))print("\n→ Late delivery is the #1 driver of 1-star reviews, regardless of freight cost")
```
